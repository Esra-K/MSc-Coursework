Command: bcftools view [OPTIONS] file.vcf.gz [REGION […​]]
<br>
<br>
Usage: View, subset and filter VCF or BCF files by position and filtering expression. Convert between VCF and BCF. Former bcftools subset.

------------------------------------

The -h tag makes the output of the view command only include headers.
<br>
So this is the last line of the output:
<br>
CHROM	POS	ID	REF	ALT	QUAL	FILTER	INFO	FORMAT	HG...
NA...


------------------------------------
Column descriptions:
<br>
CHROM: Number of chromosome of the SNPs
<br>
POS: position (index) of SNP on
chromosome
<br>
ID: SNP ID (e.g. could be found in SNPedia or DBSNP)
<br>
REF: Reference (more common)‌allele
<br>
ALT: Alternative alleles
<br>
QUAL: Quality score of SNP (measuring confidence)
<br>
INFO: additional information
<br>
FORMAT: Genotype formats
<br>
HG... or NA...
Each HG.. or NA... column represents one human sample's Genotype info.

<br>

States Conditioning:
<br>
You can increase accuracy of SHAPEIT by increasing the number of conditioning states on which haplotype estimation is based. (in HMMs, it is the length of the sequence of hidden states, e.g. the total number of days, whether sunny or rainy) By default, SHAPEIT uses 100 states per SNP across the dataset which gives good accuracy while maintaining reasonable running times. The complexity of the algorithm is linear with the number of conditioning states.
<br>
example:
```
shapeit -B gwas \
        -M genetic_map.txt \
        -O gwas.phased \
        --states 200
```
<br>
<br>
Window:
<br>
It is the length of the genotype region whose haplotypes we want to find by phasing. For GWAS dataset, we recommend a value of 2Mb, which should be fine for most application. By default, SHAPEIT2 uses windows of ~2Mb on average. Sometimes a value of 0.5Mb may give better results. To do this, use the --window option as follows:

```
shapeit -B gwas \
        -M genetic_map.txt \
        -O gwas.phased \
        --window 0.5
```

Part 7: what are genetic maps?
<br>
Heavily based on the source on part 6:
<br>

Recombination rates between SNPs are provided via a genetic map that specifies the genetic position of the SNPs in cM.
<br>
(so basically it contains the transition probabilities from one hidden state (haplotype) in one step to another hidden state in the next step, for all states and steps)

example:
```
shapeit -B gwas \
        --input-map chr20.gmap.gz \
        -O gwas.phased
```

 For the SNPs that don't have a genetic position, SHAPEIT internally determines its genetic position using linear interpolation.
<br>

If no genetic map is provided, SHAPEIT assumes a constant recombination rates between SNPs. This is unrealistic in human populations and thus may harm accuracy. The default value is 0.0004.
```
shapeit -B gwas \
        -O gwas.phased \
        --rho 0.001  # this is the constant parameter that we set
```
